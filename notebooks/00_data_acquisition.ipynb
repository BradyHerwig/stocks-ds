# Data Acquisition

Download historical stock prices with yfinance and explore the raw data.

This notebook uses the project's `stocks` package. Make sure your `.venv` kernel is selected.

```python
import sys
from pathlib import Path

# Resolve project root (works whether cwd is project root or notebooks/)
root = Path.cwd()
if root.name == "notebooks":
    root = root.parent
sys.path.insert(0, str(root / "src"))

import pandas as pd
import matplotlib.pyplot as plt

from stocks.data.loaders import download_prices, get_ticker_info
from stocks.config import DEFAULT_TICKERS
```

```python
# Download a small set of tickers with yfinance (easiest starting point)
print("Downloading sample prices...")
prices = download_prices(
    tickers=["AAPL", "MSFT", "NVDA"],
    start="2020-01-01",
    source="yfinance",
    save_raw=True,
)

print(prices.head())
print("\nIndex levels:", prices.index.names)
print("Columns:", prices.columns.tolist())
```

```python
# Basic exploration
print(
    "Date range:",
    prices.index.get_level_values("Date").min(),
    "→",
    prices.index.get_level_values("Date").max(),
)

close = prices["close"].unstack("ticker")
close.plot(figsize=(10, 5), title="Adjusted Close Prices")
plt.grid(True)
plt.show()
```

```python
# Fetch metadata for one ticker
info = get_ticker_info("AAPL")
print("AAPL info:", info)
```
