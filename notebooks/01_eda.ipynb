# Stocks EDA Notebook

This is the EDA notebook for the stock ML project.

It assumes you have already run the data pull for the example tickers (including XRP-USD) using the download script.

Data is loaded from the parquet file created in `data/raw/`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
from pathlib import Path

# Resolve project root (notebook cwd is often notebooks/)
root = Path.cwd()
if root.name == "notebooks":
    root = root.parent

data_path = root / "data" / "raw" / "prices_AAPL_MSFT_GOOGL+_2018-01-01_latest_1d.parquet"
df = pd.read_parquet(data_path)

print("Shape:", df.shape)
print("\nUnique tickers:")
print(sorted(df.index.get_level_values("ticker").unique()))
print("\nDate range:")
print(df.index.get_level_values("Date").min().date(), "to", df.index.get_level_values("Date").max().date())
```

```python
# Pivot the close prices for easier cross-ticker analysis
close = df["close"].unstack("ticker")
close.tail()
```

```python
# Plot prices for the tickers you specifically wanted (AAPL, MSFT, TSLA, NVDA + XRP)
tickers_to_plot = ["AAPL", "MSFT", "TSLA", "NVDA", "XRP-USD"]
close[tickers_to_plot].plot(title="Adjusted Close Prices (2018+)")
plt.ylabel("Price (USD)")
plt.show()
```

```python
# Daily returns
returns = close.pct_change().dropna()
returns[tickers_to_plot].plot(title="Daily Returns", alpha=0.7)
plt.ylabel("Daily Return")
plt.show()
```

```python
# Summary statistics for returns (note XRP volatility)
returns[tickers_to_plot].describe()
```

```python
# Correlation matrix of daily returns
corr = returns[tickers_to_plot].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", center=0, fmt=".2f", square=True)
plt.title("Correlation of Daily Returns")
plt.show()
```

```python
# 30-day rolling annualized volatility
rolling_vol = returns[tickers_to_plot].rolling(30).std() * (252 ** 0.5)
rolling_vol.plot(title="30-Day Rolling Annualized Volatility")
plt.ylabel("Annualized Vol")
plt.show()
```
